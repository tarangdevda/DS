```python
class atm:
    def __init__(self):
        self.menu()
        self.card_type = ''
        self.pin = ''
        self.balance = 0
        self.create_pin()
        self.check_balance = ''
        self.withdraw_money = ''


    def menu(self):
        user_input= input('''
                          Welcome to XYZ Bank ATM
                          Hello, How Can I Help You?
                          1. Press 1 To Please Select Your Card Type
                          2. Press 2 To Change The Pin
                          3. Press 3 To Deposit Money
                          4. Press 4 To Check Balance
                          5. Press 5 To Withdraw Money
                          6. Exit
                          ''')
        
        if user_input == '1':
            self.select_card_type()
            pass
        elif user_input == '2':
            self.change_pin()
            pass
        elif user_input == '3':
            self.deposit_money()
            pass
        elif user_input == '4':
            self.check_balance()
            pass
        elif user_input == '5':
            self.withdraw_money()
            pass
        else:
            exit()

    
    def select_card_type(self):
        print("Please select your card type:")
        print("1. Domestic Card")
        print("2. International Card")

        while True:
            choice = input("Enter your choice (1(Domestic) or 2(International)): ")
            if choice == "1":
                self.card_type = "Domestic"
                break
            elif choice == "2":
                self.card_type = "International"
                break
            else:
                print("Invalid input. Please enter 1 or 2 only.")

        print(f"You have selected: {self.card_type} Card\n")

    
    def create_pin(self):
        user_pin = input("Enter 4 digit number to create pin")
        self.pin = user_pin
        print("Your pin has been created")
        self.deposit_money()


    def deposit_money(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            amt = int(input("enter the amount to deposit"))
            self.balance+=amt
            print("Your money has been deposited..")
            self.menu()
        else:
            print("Entered pin is incorrect")
            self.deposit_money()

    
    def change_pin(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            new_pin = input("enter the new pin")
            self.pin = new_pin
            print("Pin updated successfully")
            self.menu()
        else:
            print("Incorrect pin")
            self.menu()

    
    def check_balance(self):
        user_pin = input("Enter your pin to check balance: ")
        if user_pin == self.pin:
            print(f"Your current balance is Rs.{self.balance}")
            self.menu()
        else:
            print("Incorrect pin")
            self.menu()

    def withdraw_money(self):
        user_pin = input("Enter your pin to withdraw money: ")
        if user_pin == self.pin:
            amt = int(input("Enter the amount to withdraw: Rs."))
            if amt <= self.balance:
                self.balance -= amt
                print(f"Rs.{amt} has been withdrawn successfully.")
                print(f"Updated balance: Rs.{self.balance}")
                self.menu()
            else:
                print("Insufficient balance.")
                self.menu()
        else:
            print("Incorrect pin")
            self.menu()

```
