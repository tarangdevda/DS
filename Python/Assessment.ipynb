```python
class atm:
    def __init__(self):
        self.card_type    = ''
        self.tax_rate     = 0.0
        self.pin          = ''
        self.balance      = 0
        self.transactions = []        
        self.create_pin()

    def menu(self):
        user_input = input('''
                          Welcome to XYZ Bank ATM
                          Hello, How Can I Help You?
                          1. Press 1 To Please Select Your Card Type
                          2. Press 2 To Change The Pin
                          3. Press 3 To Deposit Money
                          4. Press 4 To Check Balance
                          5. Press 5 To Withdraw Money
                          6. Press 6 To Mini Statement
                          7. Exit
                          ''')
        
        if user_input == '1':
            self.select_card_type()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.deposit_money()
        elif user_input == '4':
            self.check_balance()
        elif user_input == '5':
            self.withdraw_money()
        elif user_input == '6':
            self.mini_statement()
        else:
            exit()

    def select_card_type(self):
        print("Please select your card type:")
        print("1. Domestic Card")
        print("2. International Card")

        while True:
            choice = input("Enter your choice (1-Domestic / 2-International): ")
            if choice == "1":
                self.card_type = "Domestic"
                self.tax_rate  = 0.0
                break
            elif choice == "2":
                self.card_type = "International"
                self.tax_rate  = 0.05
                break
            else:
                print("Invalid input. Please enter 1 or 2 only.")

        print(f"You have selected: {self.card_type} Card\n")
        self.menu()

    def create_pin(self):
        self.pin = input("Create a 4-digit PIN: ")
        print("Your PIN has been created\n")
        self.select_card_type()      

    def change_pin(self):
        if input("Enter your current PIN: ") == self.pin:
            self.pin = input("Enter the new PIN: ")
            print("PIN updated successfully\n")
        else:
            print("Incorrect PIN\n")
        self.menu()

    def deposit_money(self):
        if input("Enter your PIN: ") == self.pin:
            amt = int(input("Enter the amount to deposit: ₹"))
            self.balance += amt
            self.transactions.append(f"Deposit : +₹{amt}")
            print("Your money has been deposited..\n")
        else:
            print("Entered PIN is incorrect\n")
        self.menu()

    def check_balance(self):
        if input("Enter your PIN to check balance: ") == self.pin:
            print(f"Your current balance is ₹{self.balance}\n")
        else:
            print("Incorrect PIN\n")
        self.menu()

    def withdraw_money(self):
        if input("Enter your PIN to withdraw money: ") != self.pin:
            print("Incorrect PIN\n")
            return self.menu()

        amt = int(input("Enter the amount to withdraw: ₹"))
        tax = amt * self.tax_rate
        total = amt + tax

        if total > self.balance:
            print("Insufficient balance.\n")
        else:
            self.balance -= total
            if self.card_type == "International":
                print(f"₹{amt} withdrawn; Tax (5 %): ₹{tax:.2f}")
                self.transactions.append(f"Withdraw   : -₹{amt} (Tax ₹{tax:.2f})")
            else:
                print(f"₹{amt} withdrawn successfully.")
                self.transactions.append(f"Withdraw   : -₹{amt}")
            print(f"Updated balance: ₹{self.balance:.2f}\n")
        self.menu()

    def mini_statement(self):
        if input("Enter your PIN to view mini statement: ") != self.pin:
            print("Incorrect PIN\n")
            return self.menu()

        print("Mini Statement")
        if not self.transactions:
            print("No transactions yet.")
        else:
            for t in self.transactions[::1][:5]:
                print(t)
        self.menu()


a = atm()
```
